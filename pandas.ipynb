Missing Data

```python
df1=pd.DataFrame({'A':[1,2,np.nan],
                 'B':[np.nan,np.nan,1],
                  'C':[1,np.nan,2]})
df1
```

```python
df1.isna()
```

```python
df1.isna().sum()
```

```python
df1.dropna() #removing of NaN values
```

```python
df1.fillna(value=2)
```

```python
df1['A'].fillna(value=df1['A'].mean())
```

```python
df1['B'].fillna(value=df1['B'].mean())
```
