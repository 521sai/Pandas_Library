#Matches played by teams

```python
matches_played=pd.concat([dataset['team1'],dataset['team2']],axis=1)
```

```python
matches_played
```

```python
team=(matches_played['team1'].value_counts() + matches_played['team2'].value_counts()).reset_index()
team
```

```python
team.columns=['team_name','played']
team
```

```python
wins=dataset['winner'].value_counts().reset_index()
wins.columns=['team_name','wins']
wins
```

```python
won=team.merge(wins, left_on='team_name', right_on='team_name', how='inner')
won['team_name']=won['team_name'].map(Teams)
won
```

```python
won['wins%']=(won['wins']/won['played'])*100
won
```

```python
sns.countplot(x='wins' ,hue='played', data=won)
```
